```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np   # обязательно добавить

# Загружаем картинку
img = Image.open("datalake.png")

# Преобразуем в массив numpy
img_array = np.array(img)

# Делаем красный канал максимальным
img_array[:, :, 0] = 255

# Показываем изображение
plt.imshow(img_array)
plt.axis('off')
plt.show()
```
